```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the new file
suggestions = pd.read_csv('tiktoksearches.csv', skiprows=1)

suggestions.columns = ['modifer_type', 'modifer', 'suggestion', 'language', 'keyword', 'search_volume', 'cost_per_click']
suggestions.head()
# Sort by search volume
top_suggestions = suggestions.sort_values('search_volume', ascending=False).head(20)

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(
    data=top_suggestions,
    y='suggestion',
    x='search_volume',
    palette='mako'
)
plt.title('Top Loafer-Related Search Suggestions (by Volume)', fontsize=16, fontweight='bold')
plt.xlabel('Search Volume')
plt.ylabel('Search Phrase')
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
def classify_query(text):
    text = text.lower()
    if 'style' in text or 'how to wear' in text or 'outfit' in text:
        return 'Styling'
    elif 'women' in text or 'female' in text or 'girls' in text:
        return 'Women'
    elif 'men' in text or 'male' in text or 'boys' in text:
        return 'Men'
    elif 'ugly' in text or 'in style' in text or 'timeless' in text:
        return 'Approval'
    elif 'wear' in text or 'suit' in text or 'worn' in text or 'style' in text or 'fit' in text or 'wears in text':
        return 'Styling'
    elif 'black' in text or 'brown' in text or 'leather' in text:
        return 'Material/Color'
    else:
        return 'Other'

# Apply the classifier
suggestions['category'] = suggestions['suggestion'].apply(classify_query)

# Plot category counts
plt.figure(figsize=(8, 5))
sns.countplot(data=suggestions, x='category', palette='Set2', order=suggestions['category'].value_counts().index)

plt.title('Themes in Loafer-Related Search Queries', fontsize=15, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Number of Queries')
plt.tight_layout()
plt.show()
```

```python
top_cpc = suggestions.sort_values('cost_per_click', ascending=False).head(15)

plt.figure(figsize=(12, 7))
sns.barplot(data=top_cpc, y='suggestion', x='cost_per_click', palette='flare')

plt.title('Most Valuable Loafer Search Queries (by CPC)', fontsize=16, fontweight='bold')
plt.xlabel('Cost Per Click ($)')
plt.ylabel('Search Phrase')
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
